### MHW statistics

```python
def mhw_stats(ds,opt,depth=0):
    mhw = {} # initialise list
    if opt=='mhw':
#         mhw['ind'] = np.where(k003['votemper'][:,depth]>=percvalu)
        mhw['ind'] = np.where(bool_mhw==1)
        mhw['time'] = ds['time_counter'].values[mhw['ind']]
        mhw['month'] = ds['time_counter.month'].values[mhw['ind']]
        mhw['temp'] = ds['votemper'][mhw['ind']]
    elif opt=='mcw':
#         mhw['ind'] = np.where(k003['votemper'][:,depth]<=percvall)
        mhw['ind'] = np.where(bool_mcw==1)
        mhw['time'] = ds['time_counter'].values[mhw['ind']]
        mhw['month'] = ds['time_counter.month'].values[mhw['ind']]
        mhw['temp'] = ds['votemper'][mhw['ind']]

    
    mhw['end'] = mhw['time'][np.insert(np.diff(mhw['ind']),-1,10)>1]
    mhw['start'] = mhw['time'][np.insert(np.diff(mhw['ind']),0,12)>1]
    mhw['duration'] = ((mhw['end']-mhw['start']).astype('timedelta64[D]'))+30
    return mhw

mhw003 = mhw_stats(k003,'mhw')
mhw004 = mhw_stats(k004,'mhw')
mhw005 = mhw_stats(k005,'mhw')
mcw003 = mhw_stats(k003,'mcw')
mcw004 = mhw_stats(k004,'mcw')
mcw005 = mhw_stats(k005,'mcw')
```

## Monte-Carlo simulation

```python
def monte_carlo_zprof(ds,var,pval):

    nvents = len(mhw003['start'])
    nmonth = len(bool_mhw[bool_mhw==1])  # number of mhw month (data points)
    mc5 = np.empty([7,len(ds.deptht)])
    mc95 = np.empty([7,len(ds.deptht)])

    for duration in np.arange(1,8):
        x=0
        n = 10000
        mc = np.empty([len(ds.deptht),n])
        while x<n:
            dummy = np.random.randint(0, len(ds.time_counter)-duration, size=(int(round(nmonth/duration)),)) # have to adjust size so total number of points is always the same
            dummy2 = dummy
            if duration>1:
                for i in np.arange(1,duration):
                    dummy2 = np.append(dummy2,dummy+i)
            mc[:,x] = ds[var][dummy2,:].mean('time_counter')
            x=x+1

        # significance shading
        mc5[duration-1] = np.percentile(mc,pval,axis=1)
        mc95[duration-1] = np.percentile(mc,100-pval,axis=1)
    return mc5,mc95
```

# Fig 5: Vertical Composites

```python
#####################################
# NNI composite vertical profiles

plt.rcParams.update({'font.size':10})
fig,ax = plt.subplots(ncols=3,figsize=(10,3.5),constrained_layout=True,sharey=True)
plt.subplots_adjust(wspace=0.05)
names = ['hindcast','thermhal','wind']
linew = 1.5
inc=2
cols = plt.get_cmap('tab10')

for ds,i,run in zip([k003,k004,k005],range(3),['k003','k004','k005']):
    mhw_nni = (ds*bool_mhw)
    mcw_nni = ds*bool_mcw
    ax[i].axvline(0,color='k',linestyle='dashed')
    
    for var,j in zip(['votemper','vosaline','vomecrty'],range(3)):
        ax[j].fill_betweenx(k003.deptht,mcl003[var][inc,:],mcu003[var][inc,:],color='darkgray',alpha=0.3,label='hindcast')
        if var=='vomecrty': depth = 'depthv'
        else: depth = 'deptht'
        mcw_nni[var].mean('time_counter').plot(ax=ax[j],yincrease=False,y=depth,label=names[i],linewidth=linew,alpha=0.7)
        mhw_nni[var].mean('time_counter').plot(ax=ax[j],yincrease=False,y=depth,linewidth=linew,alpha=0.7)
#         (deseason_month(LC[run])*bool_mhw)[var].mean(['time_counter','lat']).plot(ax=ax[j],color='blue',y=depth)
#         (deseason_month(LC[run])*bool_mcw)[var].mean(['time_counter','lat']).plot(ax=ax[j],color='blue',y=depth)
        # add LC
        if (run=='k003' and var=='vomecrty'):
            (deseason_month(LC[run])*bool_mcw)[var].mean(['time_counter','lat']).plot(ax=ax[j],color=cols(0),y=depth,
                                                                                                linestyle='dashed',linewidth=1,alpha=0.5)
            (deseason_month(LC[run])*bool_mhw)[var].mean(['time_counter','lat']).plot(ax=ax[j],color=cols(1),y=depth,
                                                                                                linestyle='dashed',linewidth=1,alpha=0.5)



# xlabels
ax[0].set_xlabel('temperature [\N{DEGREE SIGN}C]')
ax[1].set_xlabel('salinity [psu]')
ax[2].set_xlabel('meridional velocity [m/s]')
# xlimits
ax[0].set_xlim(-1,1)
ax[1].set_xlim(-0.15,0.15)
ax[2].set_xlim(-0.017,0.017)

for i in range(3): ax[i].set_ylim(800,0)

 
# two legends
lines = ax[0].get_lines()
lg1 = ax[0].legend([lines[i] for i in [2,4,6]],
                       ["hindcast", "buoyancy","wind"], loc='lower right',fontsize=9,title='Heatwaves')
lg1.get_title().set_fontsize(10)
lg2 = ax[0].legend([lines[i] for i in [1,3,5]],
                       ["hindcast", "buoyancy","wind"], loc='lower left',fontsize=9,title='Cold spells')
lg2.get_title().set_fontsize(10)

    
ax[0].add_artist(lg1)
ax[0].add_artist(lg2)
for i in range(3): ax[i].set_ylabel('')
ax[0].set_ylabel('depth [m]')

plt.suptitle('NNI composite',fontsize=12) # or plt.suptitle('Main title')

# add panel labels
for ax,i in zip(ax,range(3)) :
    ax.text(0.008, 0.9, string.ascii_lowercase[i]+')', transform=ax.transAxes, 
                size=10, weight='bold')

# plt.savefig(plotsave + 'main_figs/fig_nni_LC_composite_profiles.png',dpi=300,bbox_inches='tight')
```
