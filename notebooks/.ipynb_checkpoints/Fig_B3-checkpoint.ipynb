# Functions

```python
# 1) Subplots 3x4 for displaying seasonal evolution of section

def subplot_3x4_sec(ds,ylim,var='votemper',anomaly=False,decimals=0):
    """
    Function plots a 3x4 subplot to present seasonal cycle of sections
    
    INPUT
    ds: input data as xarray
    var: string of variable to plot
    ylim: depth range for plotting
    anomaly: boolean, i.e. True or False
    
    OUTPUT
    fig,ax: handles for figure and axes
    
    """
    
    # colormap
    if (var=='votemper' and anomaly is False): cmap = plt.get_cmap('cmo.thermal',20); label='temperature [\N{DEGREE SIGN}C]'
    elif (var=='vosaline' and anomaly is False): cmap = plt.get_cmap('viridis',20);  label='salinity [psu]'
    else: cmap = plt.get_cmap('RdBu_r',20); label = 'meridional Velocity [m/s]'
        
    # cmap bounds
    if (var=='u_normal' or var=='u_along'): anomaly=True # to have even colormap for velocity
    vmin,vmax = find_bounds(ds[var].sel(z=slice(*ylim)),decimals=decimals,anomaly=anomaly)
    print(vmin,vmax)
        
    # plotting
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    fig,ax = plt.subplots(figsize=(10,6),ncols=4,nrows=3,constrained_layout=True,sharex=True,sharey=True)
    plt.subplots_adjust(wspace=0.05)

    ii=0
    for rr in range(3):
        for cc in range(4):
            cb=ax[rr,cc].contourf(ds.lon,ds.z,ds[var][ii,:,:],levels=np.arange(-3,3.2,0.2),
                                         cmap=plt.get_cmap('RdBu_r'),vmin=np.floor(vmin),vmax=np.ceil(vmax))
#             plt.clabel(hh,fontsize=8,fmt='%2.1f')
            plt.ylim(ylim)
            ax[rr,cc].invert_yaxis()
            ax[rr,cc].set_facecolor('gray')
            cb.set_clim(np.floor(vmin),np.ceil(vmax))
            ax[rr,cc].set_title(months[ii],fontweight='bold')
#             if rr==2: ax[rr,cc].set_xlabel('longitude [\N{DEGREE SIGN}E]')
            if cc==0: ax[rr,cc].set_ylabel('depth [m]')
            ax[rr,cc].set_xticks(np.arange(110,114,1));
            ax[rr,cc].set_xticklabels(['110\N{DEGREE SIGN}E','111\N{DEGREE SIGN}E','112\N{DEGREE SIGN}E','113\N{DEGREE SIGN}E']);
            ii = ii+1
            
    # colorbar
    if (var=='votemper' and anomaly is True): label='temperature anomaly [\N{DEGREE SIGN}C]'
    elif (var=='vosaline' and anomaly is True):  label='salinity anomaly [psu]'
    elif (var=='vomecrty' and anomaly is True):  label='meridional velocity anomaly [m/s]'
    cbaxes = fig.add_axes([0.91, 0.2, 0.02, 0.6])
    cb = fig.colorbar(cb,cax=cbaxes,label=label)
#     cba = fig.colorbar(cb, ax=ax[:, cc], shrink=0.6,label=label)
    return fig, ax


#
#
######################################################################
# 2) Plot monthly profiles for seasonal cycle
def plot_sec_prof(ds,var,ylim,sec_name,anomaly=False):
    cols=plt.get_cmap('cmo.curl',12)
    months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
#     if 'LC' in sec_name: 
#         sec_prof = ds.where(ds.lon>=110,drop=True).mean('c',skipna=True)
#     else:
    sec_prof = ds.mean('c',skipna=True)
        
    matplotlib.rcParams.update({'font.size': 12})
    # plot profiles
    fig,ax = plt.subplots(figsize=(5,4))
    for i in range(12):
        sec_prof[var][i,:].plot(ax=ax,y='z',ylim=ylim,yincrease=False,color=cols(i),label=months[i])

    plt.gca().invert_yaxis()
    ax.legend(fontsize=10)
    ax.set_title(sec_name)
    ax.set_ylabel('depth [m]')
    return fig
```

```python
#################################
######### Plotting  #############
#################################
matplotlib.rcParams.update({'font.size': 9})

for sec_name in sections:

    ################## load data ###########################
    sec = xr.open_dataset('../data/section_'+ sec_name + '_' + run +'.nc')
    if (coast is True and 'LC' in sec_name): sec = sec.where(sec.lon>110,drop=True)
    elif (coast is True): sec = sec.where(sec.c<500000,drop=True)

    # derive mean seasonaly cycle
    sec_seas = sec.groupby('time_counter.month').mean('time_counter')
    sec_seas_dens = sw.pden(sec_seas['vosaline'].transpose('month','c','z')
                            ,sec_seas['votemper'].transpose('month','c','z'),
                            sec_seas.z.values,0)
    sec_seasa = sec_seas - sec_seas.mean('month')
    sec_seasa['lon']=sec['lon']


    # remove seasonal cycle
    sec_dseas = deseason_month(sec)
            
    

    ################# call plotting fucntion ##################
    # 1) subplot_3x4_sec 
    # 2) Profiles and Hovmoeller
    
    for var in variables:
        for anomaly,ds in zip([True],[sec_seasa]):
            if 'u_' in var: decimals=2
            else: decimals=1
                
            ###################################
            # 1) subplot
            matplotlib.rcParams.update({'font.size': 10})
            if 'subplot' in plot:
                fig,ax = subplot_3x4_sec(ds,[0,600],var=var,anomaly=anomaly,decimals=decimals);
                # density contours if wished
                if add_dens is True:
                    for axi,i in zip(ax.reshape(-1),range(12)):
    #                     if 'LC' in sec_name: axi.axvline(sec.c.values[np.where(sec.lon==110)[0]]/1e03,color='k',linestyle='dashed')
                        hh = axi.contour(ds.lon,sec.z,sec_seas_dens[i,::].T-1000,
                                         levels=np.arange(20,28,0.5),colors='darkred',linewidth=0.5)
                        plt.clabel(hh,fontsize=7,fmt='%2.1f')

#                     # save plot
#                     if coast is True:
#                         finished_plot(fig,fig_name=plotsave+'sec_'+sec_name+'_'+run+'_coast_seascyle_'
#                                       +var+'_anomaly_'+str(anomaly)+'_dens_contours.png')
#                     else: finished_plot(fig,fig_name=plotsave+'sec_'+sec_name+'_'+run+'_seascyle_'
#                                   +var+'_anomaly_'+str(anomaly)+'_dens_contours.png')
#                 else:
#                     # save plot
#                     if coast is True:
#                         finished_plot(fig,fig_name=plotsave+'sec_'+sec_name+'_'+run+'_coast_seascyle_'
#                                       +var+'_anomaly_'+str(anomaly)+'.png')
#                     else: finished_plot(fig,fig_name=plotsave+'sec_'+sec_name+'_'+run+'_seascyle_'
#                                   +var+'_anomaly_'+str(anomaly)+'.png')

                        
            #########################
            # 2) Profiles
            if 'profiles' in plot:
#                 fig = plot_sec_prof(ds,var,(0,600),sec_name,anomaly=anomaly);
                cols=plt.get_cmap('cmo.curl',12)
                months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
            #     if 'LC' in sec_name: 
            #         sec_prof = ds.where(ds.lon>=110,drop=True).mean('c',skipna=True)
            #     else:
                sec_prof = sec_seas.mean('c',skipna=True)
                sec_profa = sec_seasa.mean('c',skipna=True)
                matplotlib.rcParams.update({'font.size': 11})
                
                
                # plot profiles
                fig,ax = plt.subplots(ncols=2,figsize=(7,3),sharey=True,constrained_layout=True)
                for i in range(12):
                    sec_profa[var][i,:].plot(ax=ax[0],y='z',ylim=(0,600),yincrease=False,color=cols(i),label=months[i],linewidth=1)
                    sec_prof[var][i,:].plot(ax=ax[1],y='z',ylim=(0,600),yincrease=False,color=cols(i),label=months[i],linewidth=1)

                plt.gca().invert_yaxis()
                ax[1].legend(fontsize=8,loc='upper left',ncol=2)
#                 ax.set_title(sec_name)
                ax[0].set_ylabel('depth [m]')
                ax[1].set_ylabel('')
                ax[0].set_xlabel('temperature anomaly [\N{DEGREE SIGN}C]')
                ax[1].set_xlabel('temperature [\N{DEGREE SIGN}C]')
                [ax.set_title('') for ax in ax]
#                 if coast is True:
#                     finished_plot(fig,fig_name=plotsave +'sec_'+
#                                   sec_name+'_profiles_'+run+'_coast_seascyle_'
#                                   +var+'.png')
#                 else: finished_plot(fig,fig_name=plotsave + 'sec_'+
#                                     sec_name+'_profiles_'+run+'_seascyle_'
#                                   +var+'.png')                
```
