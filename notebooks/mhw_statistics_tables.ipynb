# Detect events

```python
# compare 90th percentile thresholds to std
k003std = np.std(k003['votemper'],axis=0)
percvalu,pervall = twotailperc(k003['votemper'],10)

plt.figure()
plt.plot(k003std,k003std.deptht,label='std')
plt.plot(percvalu,k003std.deptht,label='90th perc')
plt.gca().legend()
plt.gca().invert_yaxis()
```

```python


def detect_events(pval,opt='const'):

    if opt=='const':
        # detect events: constant threshold
        # significance upper, lower level
        percvalu,percvall = twotailperc(k003['votemper'][:,0],pval)
        print('calculate constant threshold')

    
    elif opt=='seas':
        # seasonal,monthly threshold
        percvalu = np.empty([4,])
        percvall = np.empty([4,])
        dummy = k003.time_counter.dt.season
        for season,i in zip(['DJF','MAM','JJA','SON'],range(4)):
            test = k003.votemper[k003.time_counter.dt.season==season,0]
            percvalu[i] = np.nanpercentile(test,100-pval,axis=0)
            percvall[i] = np.nanpercentile(test,pval,axis=0)
            dummy[dummy==season]=percvalu[i]
            dummy=dummy.astype(np.float)

    elif opt=='monthly' :       
        percvalu = np.empty([12,])
        percvall = np.empty([12,])
        for month in np.arange(0,12):
            test = k003.votemper[k003.time_counter.dt.month==month+1,0]
            percvalu[month] = np.nanpercentile(test,100-pval,axis=0)
            percvall[month] = np.nanpercentile(test,pval,axis=0)

        # multiply seasonal cycle to match timeseries
        timelen = len(np.unique(k003.time_counter.dt.year))
        percvalu = np.tile(percvalu,timelen)
        percvall = np.tile(percvall,timelen)
        print('calculate monthly threshold')

    # create boolean-type vector to separate positive and negative events
    bool_mhw = (k003['votemper'][:,0]-percvalu)
    bool_mhw[bool_mhw>=0] = 1
    bool_mhw[bool_mhw<0] = np.NaN

    bool_mcw = (k003['votemper'][:,0]-percvall)
    bool_mcw[bool_mcw>0] = np.NaN
    bool_mcw[bool_mcw<=0] = 1
    
    return bool_mhw,bool_mcw

def mhw_stats(pval,ds,opt):
    bool_mhw,bool_mcw = detect_events(pval,opt)


    ##################################
    # Marine Heatwave & cold spell statistics
    ##################################

    def mhw_stats(ds,opt,depth=0):
        mhw = {} # initialise list
        
        if opt=='mhw':
    #         mhw['ind'] = np.where(k003['votemper'][:,depth]>=percvalu)
            # manually separate the long 2-year event (2010-2012)
            test = bool_mhw.copy(True)
            test[np.where((bool_mhw.time_counter.dt.year==2011) & (bool_mhw.time_counter.dt.month==8))] = 0
    
            mhw['ind'] = np.where(test==1)
            mhw['time'] = ds['time_counter'].values[mhw['ind']]
            mhw['month'] = ds['time_counter.month'].values[mhw['ind']]
            mhw['temp'] = ds['votemper'][mhw['ind']]
            mhw['sal'] = ds['vosaline'][mhw['ind']]
            mhw['vvel'] = ds['vomecrty'][mhw['ind']]
            mhw['mld'] = ds['somxl010'][mhw['ind']]
            mhw['end'] = mhw['time'][np.insert(np.diff(mhw['ind']),-1,10)>1]
            mhw['start'] = mhw['time'][np.insert(np.diff(mhw['ind']),0,12)>1]
            mhw['oni'] = (indices['oni'][mhw['ind']])            
            # have to add August back into one of the events (will do end of 2011 event)
            # it's the 4th last event 
            mhw['end'][-4] = mhw['end'][-4] + np.timedelta64(31,'D')  # July has 31 days, so need to add 31
            # check that it worked
            print(mhw['end'][-4])
            mhw['duration'] = (((mhw['end']-mhw['start'])+ np.timedelta64(31,'D')))/30
            mhw['cwi'] = (cwi_tau[mhw['ind']]*(1e02))
            
            # derive additional diagnostics
            cumint = np.empty([23,len(mhw['duration'])])
            maxint = np.empty([23,len(mhw['duration'])])
            maxsal = np.empty([23,len(mhw['duration'])])
            cwi = np.empty([len(mhw['duration'])])
            cwi_max = np.empty([len(mhw['duration'])])
            cwi_min = np.empty([len(mhw['duration'])])
            oni = np.empty([len(mhw['duration'])])

            
            for i in range(len(mhw['duration'])):
                cumint[:,i] = sum(mhw['temp'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])))
                maxint[:,i] = np.max(mhw['temp'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])))
                maxsal[:,i] = np.max(abs(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))))
                dummy = np.argmax(abs(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))),axis=0)
                sign = np.sign(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))[dummy,:])
                maxsal[:,i] = maxsal[:,i]*sign
                # coastal wind index
                cwi[i] = mhw['cwi'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])).mean('time_counter')
                cwi_max[i] = mhw['cwi'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])).max('time_counter')
                cwi_min[i] = mhw['cwi'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])).min('time_counter')
                oni[i] = indices['oni'].sel(Month=slice(mhw['start'][i]-np.timedelta64(17,'D'),
                                             mhw['end'][i])).mean('Month').round(2).values
            mhw['cumint'] = cumint
            mhw['maxint'] = maxint
            mhw['maxsal'] = maxsal
            mhw['cwi'] = cwi
            mhw['oni'] = oni
                
        elif opt=='mcw':
    #         mhw['ind'] = np.where(k003['votemper'][:,depth]<=percvall)
            mhw['ind'] = np.where(bool_mcw==1)
            mhw['time'] = ds['time_counter'].values[mhw['ind']]
            mhw['month'] = ds['time_counter.month'].values[mhw['ind']]
            mhw['temp'] = ds['votemper'][mhw['ind']]
            mhw['sal'] = ds['vosaline'][mhw['ind']]
            mhw['vvel'] = ds['vomecrty'][mhw['ind']]
            mhw['mld'] = ds['somxl010'][mhw['ind']]
            mhw['end'] = mhw['time'][np.insert(np.diff(mhw['ind']),-1,10)>1]
            mhw['start'] = mhw['time'][np.insert(np.diff(mhw['ind']),0,12)>1]
            mhw['duration'] = (((mhw['end']-mhw['start'])+ np.timedelta64(31,'D')))/30
            mhw['cwi'] = (cwi_tau[mhw['ind']]*(1e02))
            mhw['oni'] = (indices['oni'][mhw['ind']])            
            cumint = np.empty([23,len(mhw['duration'])])
            maxint = np.empty([23,len(mhw['duration'])])
            maxsal = np.empty([23,len(mhw['duration'])])
            oni = np.empty([len(mhw['duration'])])
            cwi = np.empty([len(mhw['duration'])])
            cwi_max = np.empty([len(mhw['duration'])])
            cwi_min = np.empty([len(mhw['duration'])])
            
            for i in range(len(mhw['duration'])):
                cumint[:,i] = sum(mhw['temp'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])))
                maxint[:,i] = np.min(mhw['temp'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])))
                maxsal[:,i] = np.max(abs(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))))
                dummy = np.argmax(abs(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))),axis=0)
                sign = np.sign(mhw['sal'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i]))[dummy,:])
                maxsal[:,i] = maxsal[:,i]*sign
                cwi[i] = mhw['cwi'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])).mean('time_counter')
                cwi_max[i] = mhw['cwi'].sel(time_counter=slice(mhw['start'][i],mhw['end'][i])).max('time_counter')
                oni[i] = indices['oni'].sel(Month=slice(mhw['start'][i]-np.timedelta64(17,'D'),
                                             mhw['end'][i])).mean('Month').round(2).values                
            mhw['cumint'] = cumint
            mhw['maxint'] = maxint
            mhw['maxsal'] = maxsal
            mhw['cwi'] = cwi
            mhw['oni'] = oni

        return mhw


    # call function
    mhw003 = mhw_stats(ds,'mhw')
#     mhw004 = mhw_stats(k004,'mhw')
#     mhw005 = mhw_stats(k005,'mhw')
    mcw003 = mhw_stats(ds,'mcw')
#     mcw004 = mhw_stats(k004,'mcw')
#     mcw005 = mhw_stats(k005,'mcw')

    # How many events detected?
    print('Detected MHW events: '+ np.str(len(mhw003['duration'])) +' \n')
    print('Detected Cold spell events: '+ np.str(len(mcw003['duration'])) +' \n')

    # create data frame with information
    pd_mhw = pd.DataFrame({'Event':np.arange(1,len(mhw003['start'])+1), 'Start':np.array(mhw003['start'].astype('datetime64[M]')),
              'End':mhw003['end'].astype('datetime64[M]'),'dummy':mhw003['duration'],
              'Intensity [$^\circ$C]':mhw003['maxint'][0,:].round(2),'Salinity [psu]':mhw003['maxsal'][0,:].round(2),
              'ONI':mhw003['oni'].round(2),'CWI':mhw003['cwi'].round(2)})
    pd_mcw = pd.DataFrame({'Event':np.arange(1,len(mcw003['start'])+1), 'Start':mcw003['start'].astype('datetime64[M]'),
              'End':mcw003['end'].astype('datetime64[M]'), 'dummy':mcw003['duration'],
              'Intensity [$^\circ$C]':mcw003['maxint'][0,:].round(2),'Salinity [psu]':mcw003['maxsal'][0,:].round(2),
              'ONI':mcw003['oni'].round(2),'CWI':mcw003['cwi'].round(2)})
    # adjust duration format to month
    pd_mhw["Duration"] = ""
    pd_mcw["Duration"] = ""
    for i in range(len(mhw003['duration'])):
        pd_mhw["Duration"][i] = str(pd_mhw['dummy'][i].round('D')).replace('days 00:00:00','month')
    for i in range(len(mcw003['duration'])):
        pd_mcw["Duration"][i] = str(pd_mcw['dummy'][i].round('D')).replace('days 00:00:00','month')
    
    
    return pd_mhw.drop(['dummy'],axis=1),pd_mcw.drop(['dummy'],axis=1),mhw003,mcw003

    
##################################
# call function
##################################
pval=10
k003stats_mhw,k003stats_mcw,mhw003,mcw003 = mhw_stats(pval,k003,'monthly')
_,_,mhw004,mcw004 = mhw_stats(pval,k004,'monthly')
_,_,mhw005,mcw005 = mhw_stats(pval,k005,'monthly')

# k003stats_mhw,k003stats_mcw,mhw003,mcw003 = mhw_stats(pval,k005,'monthly')
# adjust time format
k003stats_mhw['Start'] = k003stats_mhw['Start'].dt.strftime('%Y-%b')
k003stats_mhw['End'] = k003stats_mhw['End'].dt.strftime('%Y-%b')
k003stats_mcw['Start'] = k003stats_mcw['Start'].dt.strftime('%Y-%b')
k003stats_mcw['End'] = k003stats_mcw['End'].dt.strftime('%Y-%b')

cols=['Event','Start','End','Duration','Intensity [$^\circ$C]','Salinity [psu]','ONI','CWI']
k003stats_mhw=k003stats_mhw.ix[:, cols]
k003stats_mcw=k003stats_mcw.ix[:, cols]



bool_mhw,bool_mcw = detect_events(pval,'monthly')
```
